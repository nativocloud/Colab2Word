```python
import pandas as pd
import matplotlib.pyplot as plt
from docx.enum.section import WD_ORIENT
```

```python
from colab2word import DocumentGenerator, DocumentSection
```

```python

# Create test data
sales_data = {
    'Product': ['Laptop', 'Tablet', 'Phone', 'Watch'],
    'Sales': [1200, 800, 1500, 300],
    'Revenue': [1200000, 400000, 750000, 90000],
    'Rating': [4.8, 4.5, 4.9, 4.2]
}
df = pd.DataFrame(sales_data)

# Create test plot
plt.figure(figsize=(10, 6))
plt.bar(df['Product'], df['Sales'], color='skyblue')
plt.title('Product Sales Distribution')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.grid(True, alpha=0.3)
plot_figure = plt.gcf()

# Initialize DocumentGenerator with display enabled
doc_gen = DocumentGenerator(
    output_dir='test_output',
    default_filename='display_test.docx',
    display_output=True
)

# Create test sections
sections = [
    # Text section
    DocumentSection(
        content="""
# Sales Performance Overview

Our product line shows strong performance across all categories:
- Laptops remain our highest revenue generator
- Phones lead in unit sales
- All products maintain above 4.0 rating
        """,
        header="Executive Summary",
        description="Overview of Q1 2024 Sales Performance"
    ),
    
    # Table section
    DocumentSection(
        content=df,
        header="Sales Metrics",
        description="Detailed breakdown of sales performance by product"
    ),
    
    # Plot section
    DocumentSection(
        content=plot_figure,
        header="Sales Distribution",
        description="Visual comparison of unit sales across product lines"
    )
]

# Generate document
output_path = doc_gen.generate_document(
    sections=sections,
    title="Sales Performance Report",
    author="Test User",
    create_new=True
)

print(f"\nDocument generated at: {output_path}")
```

## Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from colab2word import DocumentGenerator, DocumentSection

# Create sample data
np.random.seed(42)
n_samples = 100

# Create a complex dataset
data = pd.DataFrame({
    'Category': np.repeat(['A', 'B', 'C', 'D'], n_samples//4),
    'Value1': np.random.normal(100, 15, n_samples),
    'Value2': np.random.normal(75, 25, n_samples),
    'Timestamp': pd.date_range(start='2024-01-01', periods=n_samples, freq='D'),
    'Group': np.random.choice(['Group1', 'Group2', 'Group3'], n_samples),
    'Score': np.random.uniform(0, 100, n_samples)
})

# Create multiple visualizations
# 1. Complex Distribution Plot
plt.figure(figsize=(12, 8))
sns.violinplot(data=data, x='Category', y='Value1', hue='Group')
plt.title('Distribution by Category and Group')
violin_plot = plt.gcf()

# 2. Time Series with multiple groups
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Timestamp', y='Value2', hue='Group', style='Group')
plt.title('Time Series Analysis by Group')
plt.xticks(rotation=45)
time_series = plt.gcf()

# 3. Joint Plot
joint_plot = sns.jointplot(
    data=data,
    x='Value1',
    y='Value2',
    hue='Group',
    kind='scatter',
    height=8
)

# Initialize document generator
doc_gen = DocumentGenerator(
    output_dir='output',
    default_filename='seaborn_analysis.docx',
    display_output=True
)

# Create sections
sections = [
    DocumentSection(
        content="# Advanced Data Visualization Analysis\n\nThis report demonstrates complex "
                "statistical visualizations using Seaborn.",
        content_type="text",
        header="Data Analysis Report",
        description="Analysis of multivariate relationships and distributions"
    ),
    
    DocumentSection(
        content=violin_plot,
        content_type="plot",
        header="Distribution Analysis",
        description="Violin plot showing value distributions across categories and groups",
        style_options={'width': 8}
    ),
    
    DocumentSection(
        content=time_series,
        content_type="plot",
        header="Time Series Analysis",
        description="Temporal trends across different groups",
        style_options={'width': 8},
        page_break_before=True
    ),
    
    DocumentSection(
        content=joint_plot.figure,
        content_type="plot",
        header="Correlation Analysis",
        description="Joint distribution of Value1 and Value2 with group distributions",
        style_options={'width': 8},
        page_break_before=True
    )
]

# Generate document
output_path = doc_gen.generate_document(
    sections=sections,
    title="Advanced Visualization Report",
    author="Seaborn Analysis",
    create_new=True
    
)

print(f"\nDocument generated at: {output_path}")
```

```python
# Create test plots and document
import matplotlib.pyplot as plt
import numpy as np

# Create portrait plot
plt.figure(figsize=(8, 6))
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.title('Sine Wave')
portrait_plot = plt.gcf()

# Create landscape plot
plt.figure(figsize=(12, 6))
plt.plot(x, np.cos(x))
plt.title('Cosine Wave')
landscape_plot = plt.gcf()

# Initialize DocumentGenerator
doc_gen = DocumentGenerator(
    output_dir='test_output',
    default_filename='plots_test.docx',
    display_output=True
)

# Create sections with different orientations and comments
sections = [
    DocumentSection(
        content=portrait_plot,
        content_type='plot',
        header="Portrait Plot",
        description="A simple sine wave plot",
        comments="This plot uses portrait orientation and shows a single period of a sine wave.",
        orientation='portrait',
        style_options={'width': 6}
    ),
    DocumentSection(
        content=landscape_plot,
        content_type='plot',
        header="Landscape Plot",
        description="A simple cosine wave plot",
        comments="This plot uses landscape orientation for better visibility of the wide data.",
        orientation='landscape',
        style_options={'width': 8}
    )
]

# Generate document
doc_gen.generate_document(
    sections=sections,
    title="Plot Orientation Test",
    author="Data Science Team",
    create_new=True
)
```
